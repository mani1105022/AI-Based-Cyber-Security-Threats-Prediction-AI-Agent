# ***Task:Employee Salary Analysis***

Prompt:Load the employee dataset

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("employee_dataset.csv")

# Display dataset
df
```

Prompt:Show basic info of dataset

```python
df.info()
```

Prompt : Calculate Total Salary for Each Employee

```python
df["TotalSalary"] = df["HourlyRate"] * df["HoursWorked"]
df
```

Prompt : Find the Highest Paid Employee

```python
highest_paid = df.loc[df["TotalSalary"].idxmax()]
highest_paid
```

Prompt : Show Summary Statistics

```python
df.describe()
```

Prompt: Visualize Employee Total Salary (Bar Chart)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["TotalSalary"])
plt.title("Total Salary of Employees")
plt.xlabel("Employee Name")
plt.ylabel("Total Salary (INR)")
plt.show()
```

Prompt : Visualize Hours Worked (Pie Chart)

```python
plt.figure(figsize=(6,6))
plt.pie(df["HoursWorked"], labels=df["Name"], autopct="%1.1f%%")
plt.title("Hours Worked Distribution")
plt.show()
```
